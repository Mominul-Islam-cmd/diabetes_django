```python
df=df.drop(columns=['smoking_history'])
df
```

```python
df1=pd.get_dummies(df['gender'])
df1
```

```python
# Ensure the dummy variables are numeric
df1= df1.astype(np.int64)
```

```python
df1.value_counts()
```

```python
n_df = pd.concat([df, df1], axis=1).reindex(df.index)
n_df
```

```python
df=n_df.drop(columns=['gender'])
df
```

```python
df.info()
```

```python
x=df.drop(columns='diabetes',axis=1)
y=df['diabetes']
```

# DataSet Split (train_test Split)

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
xtrain
```

```python
# Standardize the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(xtrain)
X_test_scaled = scaler.transform(xtest)
```

```python
X_train_scaled
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
LoR=LogisticRegression()
```

```python
LoR.fit(X_train_scaled,ytrain)
```

```python
LoR.score(X_test_scaled,ytest)
```

```python
pred=LoR.predict(X_test_scaled)
pred
```

```python
from sklearn.metrics import classification_report
```

```python
lor_cr=classification_report(ytest,pred)
print(lor_cr)
```

# #  KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor

knn = KNeighborsClassifier(n_neighbors=2) 
knn.fit(X_train_scaled, ytrain)
```

```python
knn.score(X_test_scaled,ytest)
```

```python
knnpred=LoR.predict(X_test_scaled)
knnpred
```

```python
knn_cr=classification_report(ytest,knnpred)
print(knn_cr)
```

```python
# Function to accept user input
def get_user_input():
    input_data = []
    feature_names = x.columns
    for feature in feature_names:
        value = float(input(f'Enter value for {feature}: '))
        input_data.append(value)
    return input_data

user_input = get_user_input()
print(f'User input: {user_input}')
```

```python
# Transform the user input using the same scaler
user_input_r = np.array(user_input).reshape(1, -1)
user_input_r

```

```python
user_input_s = scaler.transform(user_input_r)
user_input_s
```

```python
# Make prediction
prediction = knn.predict(user_input_s)
print(f'Prediction: {"Diabetes" if prediction[0] == 1 else "No Diabetes"}')
```
